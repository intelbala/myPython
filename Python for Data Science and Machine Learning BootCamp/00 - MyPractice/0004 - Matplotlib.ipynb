```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

#### Functional Method

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
#plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

#### Object Oriented method 

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])   #List elements: [left, bottom, width, height]
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Plot Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title("Larger plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
# Subplot using OOP
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,.8,0.8])
axes.plot(x,y)
```

## Customizing plot

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
#axes.plot(x,y, color='green')      #direct colors
#axes.plot(x,y, color='#FF8C00',linewidth=3, alpha = 0.5)     #RGB HEX Code, Alpha = Transparency, linewidth or lw
axes.plot(x,y, color='#FF8C00',linewidth=1, linestyle='--',marker='o',markersize=10,
         markerfacecolor='yellow',markeredgewidth=3, markeredgecolor='green') #markers are used to plot points
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y, color='purple',linewidth=1, linestyle='--')
axes.set_xlim([0,5])
axes.set_ylim([0,30])
```
