### Distribution plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)  #bins is interval
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex')   #creates a joint plot for all numerical variables
```

```python
sns.rugplot(tips['total_bill'])  #counts the number of elements within a bin and adds a dash. So it shows the data concentration
```

```python
sns.distplot(tips['total_bill'])
```

### Categorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=sum) #estimator is an aggregate function
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)  #standard deviation
```

### Matrix plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#In order for Heat map to work properly, the data should be in matrix form
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linewidths=1,linecolor='white')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

### Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

### Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot('total_bill','tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,height=8)
```
