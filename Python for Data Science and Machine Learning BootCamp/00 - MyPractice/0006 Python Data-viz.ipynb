### Pandas built in data visualization

```python
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df1['A'].hist(bins=20)
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
df2.plot.area(alpha=0.4)  #alpha parameter adds some transparency
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=30)
```

```python
df1.plot.line(y='B',figsize=(10,2))
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.box()
```
