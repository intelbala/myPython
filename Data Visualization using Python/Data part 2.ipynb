### Merge

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
```

```python
dframe1
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'],'data_set_2':np.arange(3)})
```

```python
dframe2
```

```python
pd.merge(dframe1,dframe2)  #Merge based on key
```

```python
pd.merge(dframe1,dframe2,on='key')
```

```python
pd.merge(dframe1,dframe2,on='key',how='left')  #similar to left join
```

```python
pd.merge(dframe1,dframe2,on='key',how='right')  #similar to right join
```

```python
pd.merge(dframe1,dframe2,on='key',how='outer')  #similar to outer join
```

```python
#Many to Many joins
```

```python
dframe3 = DataFrame({'key':['X','X','X','Y','Z','Z'],'data_set_3':range(6)})
```

```python
dframe4 = DataFrame({'key':['Y','Y','X','X','Z'],'data_set_4':range(5)})
```

```python
dframe3
```

```python
dframe4
```

```python
pd.merge(dframe3,dframe4)
```

### Merge on index

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
df_left = pd.read_csv('bb_merge_index_1.csv')
```

```python
df_left
```

```python
df_right = pd.read_csv('bb_merge_index_2.csv',index_col=1)
```

```python
df_right
```

```python
pd.merge(df_left,df_right,left_on='key',right_index=True)
#left_on - left dataframe will merge on key
#right_index - Right dataframe will merge on index
```

```python
df_left_hr = DataFrame({'key1':['SF','SF','SF','LA','LA'],
                        'key2':[10,20,30,20,30],
                       'data_set':np.arange(5.)})
```

```python
df_left_hr
```

```python
df_right_hr = DataFrame(np.arange(10).reshape(5,2),
                        index=[['LA','LA','SF','SF','SF'],
                               [20,10,10,10,20]],
                               columns=['col1','col2'])
```

```python
df_right_hr
```

```python
pd.merge(df_left_hr,df_right_hr,left_on=['key1','key2'],right_index=True,how='outer')
```
